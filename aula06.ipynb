<h2><b>Sinal Contínuo</b></h2>

```python
import numpy as np
import matplotlib.pyplot as plt

tempo_v = np.linspace(0, 1, 1000)
freq_sinal = 20
sinal_cont = np.sin(2*np.pi*freq_sinal*tempo_v)
```

<h5><em>Transformada de Fourier</em></h5>

```python
frequen = np.fft.fftfreq(len(tempo_v), tempo_v[1] - tempo_v[0])
```

<h5><em>Filtragem</b></h5>

```python
ruido = 0.5 * np.random.randn(len(tempo_v))
sinal_com_ruido = sinal_cont + ruido

freq_cort = 15
transform = np.fft.fft(sinal_com_ruido)
transform[frequen > freq_cort] = 0
sinal_filter = np.fft.ifft(transform)
```

<h5><em>Análise Espectral</em></h5>

```python
espectro = np.abs(transform)
```

<h1>PLOT</h1>

```python
plt.figure(figsize=(8, 6))
plt.plot(frequen, espectro)
plt.plot(tempo_v, sinal_cont, label="Sinal Original")
plt.plot(tempo_v, sinal_com_ruido, 'r', label="Sinal com Ruído")
plt.plot(tempo_v, sinal_filter, 'g', label="Sinal Filtrado")
plt.legend()
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.title('Sinal')
plt.grid(True)
plt.show()
```

<h1><b>Aperfeiçoamento das imagens</b></h1>

<h5><em>Imagem 1</em></h5>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./imgsaula06/1.png')

imagem_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_sem_ruido = cv2.medianBlur(imagem_gray, 5)

plt.figure(figsize=(6, 4))
plt.imshow(img, cmap="gray")
plt.title('Antes Imagem 1')
plt.axis('off')
plt.show()

plt.figure(figsize=(6, 4))
plt.imshow(img_sem_ruido, cmap="gray")
plt.title('Depois Imagem 1')
plt.axis('off')
plt.show()
```

<h5><em>Imagem 2</em></h5>

```python
img2 = cv2.imread('./imgsaula06/2.png')

img_filter = cv2.GaussianBlur(img2, (5, 5), 0)
imagem_normalizada = cv2.normalize(img_filter, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

plt.figure(figsize=(5, 3))
plt.imshow(img2, cmap="gray")
plt.title('Antes Imagem 2')
plt.axis('off')
plt.show()

plt.figure(figsize=(5, 3))
plt.imshow(imagem_normalizada, cmap="gray")
plt.title('Depois Imagem 2')
plt.axis('off')
plt.show()
```

<h5><em>Imagem 3</em></h5>

```python
img3 = cv2.imread('./imgsaula06/3.png')

img_gray = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
img_filter = cv2.GaussianBlur(img_gray, (5, 5), 0)
img_equalize = cv2.equalizeHist(img_filter)
img_sem_ruido = cv2.medianBlur(img_equalize, 3)

plt.figure(figsize=(6, 4))
plt.imshow(img3, cmap="gray")
plt.title('Antes Imagem 3')
plt.axis('off')
plt.show()

plt.figure(figsize=(6, 4))
plt.imshow(img_sem_ruido, cmap="gray")
plt.title('Depois Imagem 3')
plt.axis('off')
plt.show()
```
